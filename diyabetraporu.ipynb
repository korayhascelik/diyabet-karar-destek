            Diyabet (Tip 2), dünya çapında milyonlarca insanı etkileyen, kardiyovasküler hastalıklar, böbrek yetmezliği ve görme kaybı gibi ciddi komplikasyonlara yol açabilen kronik bir metabolik hastalıktır. Geleneksel tıp yaklaşımları genellikle hastalık oluştuktan sonra tedaviye (reaktif) odaklanırken, modern sağlık vizyonu hastalığı <b>pre-diyabet</b> (gizli şeker) aşamasında yakalamayı (proaktif) hedefler. Bu projenin ana konusu, bireylerin klinik ve demografik özniteliklerini kullanarak makine öğrenmesi algoritmalarıyla diyabetik olma risklerini tahmin etmektir.
        </p>
        <h4 style="color: #7DD3FC; margin-bottom: 5px; margin-top: 20px;">🏥 Sağlık Bilişimi ve Stratejik Önem</h4>
        <p style="color: #CBD5E1; font-size: 15px; line-height: 1.6; font-family: 'Arial', sans-serif;">
            Bu problem sıradan bir sınıflandırma (classification) ödevi değil, doğrudan bir <b>Klinik Karar Destek Sistemi (CDSS)</b> mimarisidir. Stratejik önemi üç boyutta incelenebilir:
            <ul style="color: #CBD5E1; font-size: 15px; line-height: 1.6;">
                <li><b>Bireysel Boyut:</b> Erken uyarı sistemi sayesinde bireylerin yaşam tarzı (diyet, egzersiz) değişiklikleriyle hastalığın önüne geçilmesi.</li>
                <li><b>Kurumsal Boyut:</b> Sağlık kuruluşlarının poliklinik yığılmalarını önlemesi, kronik tedavi ve yoğun bakım maliyetlerini minimize etmesi.</li>
                <li><b>Karar Destek Boyutu:</b> Hekimlerin teşhis koyarken gözden kaçırabileceği doğrusal olmayan çoklu klinik sinyalleri (örn: insülin, glikoz ve yaş kombinasyonları) bir yapay zeka süzgecinden geçirerek ikinci bir görüş (second opinion) sunması.</li>
            </ul>
        </p>
        <h4 style="color: #7DD3FC; margin-bottom: 5px; margin-top: 20px;">🎯 Veri Madenciliği Hedefleri (Analitik Vizyon)</h4>

            Projenin teknik hedefi, yalnızca "yüksek doğruluk (accuracy)" elde eden bir model kurmak <b>değildir.</b> Tıbbi tahmin problemlerinde asıl risk <b>False Negative (Tip 2 Hata)</b> oranıdır. Yani diyabet riski taşıyan bir hastaya "sağlıklısın" diyerek eve göndermek, sağlıklı birine "risklisin" demekten (False Positive) çok daha büyük bir tehlikedir. Bu nedenle projede ana hedef; <b>Recall (Duyarlılık / Sensitivity)</b> skorunu maksimize ederken, makul bir Precision dengesi kurmaktır.

        <p style="color: #CBD5E1; font-size: 15px; line-height: 1.6; font-family: 'Arial', sans-serif;">
            Bu projede, sağlık veri madenciliğinde bir referans (benchmark) olan <b>Pima Indians Diabetes Database</b> kullanılmıştır. Veri seti 768 hastaya (gözlem) ait 8 klinik/demografik bağımsız değişken (Pregnancies, Glucose, BloodPressure, SkinThickness, Insulin, BMI, DiabetesPedigreeFunction, Age) ve 1 hedef değişken (Outcome) içermektedir. Hedef değişken, hastanın diyabetik (1) veya sağlıklı (0) olduğunu gösterir.
        </p>
        <h4 style="color: #34D399; margin-bottom: 5px; margin-top: 20px;">⚠️ Mantıksal Tutarsızlıklar ve "Sıfır" Problemi</h4>
        <p style="color: #CBD5E1; font-size: 15px; line-height: 1.6; font-family: 'Arial', sans-serif;">
            Ham veri incelendiğinde <code>Glucose, BloodPressure, SkinThickness, Insulin, BMI</code> kolonlarında "0" (sıfır) değerleri olduğu tespit edilmiştir. İnsan biyolojisi gereği kan basıncı, vücut kitle indeksi veya kan şekerinin sıfır olması ölümcül/imkansız bir durumdur. Bu değerler gerçek birer sıfır değil, sisteme girilmeyen (Missing/NaN) değerlerdir. Modellerin "yanlış öğrenmesini (bias)" engellemek için bu değerler <b>NaN</b> olarak işaretlenmiş ve aykırı değerlerden (outliers) daha az etkilenmesi sebebiyle medyan (ortanca) yöntemi ile doldurulmuştur (Imputation).
        </p>
        <h4 style="color: #34D399; margin-bottom: 5px; margin-top: 20px;">🛠️ Öznitelik Mühendisliği (Feature Engineering)</h4>
        <p style="color: #CBD5E1; font-size: 15px; line-height: 1.6; font-family: 'Arial', sans-serif;">
            Makine öğrenmesi algoritmaları doğrusal olmayan (non-linear) yapıları öğrenmede zorlanabilir. Bu nedenle klinik eşik değerleri (threshold) baz alınarak yeni özellikler türetilmiştir:
            <ul style="color: #CBD5E1; font-size: 15px; line-height: 1.6;">
                <li><b>BMI_Category:</b> Dünya Sağlık Örgütü (WHO) standartlarına göre; Zayıf, Normal, Kilolu ve Obez segmentasyonu.</li>
                <li><b>Glucose_Risk:</b> Şeker düzeyine göre; Normal, Pre-diyabetik ve Diyabetik kategorizasyonu.</li>
                <li><b>Age_Group:</b> Yaşın metabolizma üzerindeki etkisini modellemek için; Genç Yetişkin, Orta Yaşlı ve İleri Yaş gruplandırması.</li>
            </ul>
            Kategorik özellikler daha sonra <i>One-Hot Encoding</i> ile algoritmaların anlayacağı sayısal formata dönüştürülmüştür. İşlenen nihai veri, klasör hiyerarşimize uygun olarak <code>data/processed</code> dizinine kaydedilmiştir.
        </p>
    </div>
</div>

```python
import pandas as pd
import numpy as np
import plotly.express as px
import plotly.graph_objects as go
import plotly.io as pio
import os

# Plotly için Koyu Temayı (Dark Mode) varsayılan yapıyoruz
pio.templates.default = "plotly_dark"

# 1. Ham Veriyi Yükleme
df = pd.read_csv('data/raw/diabetes.csv')

print("🛠️ Veri Ön İşleme (Data Preprocessing) Başlıyor...\n")

# --- GÖRSELLEŞTİRME 1: Hedef Değişken (Sınıf Dengesi) ---
fig1 = px.pie(df, names='Outcome', title="Bağımlı Değişken Sınıf Dağılımı (Diyabetik vs Sağlıklı)",
              color_discrete_sequence=['#3B82F6', '#EF4444'], hole=0.5)
fig1.update_traces(textinfo='percent+label', textfont_size=14)
fig1.update_layout(paper_bgcolor='#0F172A', plot_bgcolor='#0F172A', font_color='#CBD5E1')
fig1.show()
# Grafiği figures klasörüne kaydet
fig1.write_html("figures/class_distribution.html")

# 2. Mantıksal Tutarsızlıkları Çözme (0'ları NaN yapıp Medyan ile doldurma)
cols_missing = ['Glucose', 'BloodPressure', 'SkinThickness', 'Insulin', 'BMI']

print("🔍 Biyolojik olarak imkansız '0' değerleri tespit ediliyor:")
for col in cols_missing:
    zero_count = (df[col] == 0).sum()
    print(f"   - {col}: {zero_count} adet '0' değeri bulundu.")

# 0'ları NaN yap
df[cols_missing] = df[cols_missing].replace(0, np.nan)

# Eksik verileri, sınıfların (Outcome 0 ve 1) kendi içindeki medyanlarına göre doldurmak daha tıbbi bir yaklaşımdır.
for col in cols_missing:
    df[col] = df.groupby('Outcome')[col].transform(lambda x: x.fillna(x.median()))

print("\n✅ Tüm '0' değerleri sınıfların kendi medyan (ortanca) değerleri ile dolduruldu.")

# 3. Öznitelik Mühendisliği (Feature Engineering)
# a. Glikoz Risk Kategorisi
df['Glucose_Risk'] = pd.cut(df['Glucose'], bins=[0, 139, 199, 500], labels=['Normal', 'Prediabetes', 'Diabetes'])

# b. BMI (Vücut Kitle İndeksi) Kategorisi
df['BMI_Category'] = pd.cut(df['BMI'], bins=[0, 18.5, 24.9, 29.9, 100], labels=['Underweight', 'Normal', 'Overweight', 'Obese'])

# c. Yaş Grubu Segmentasyonu
df['Age_Group'] = pd.cut(df['Age'], bins=[20, 35, 50, 100], labels=['Young_Adult', 'Middle_Aged', 'Senior'])

# --- GÖRSELLEŞTİRME 2: Glikoz Riski ve Diyabet İlişkisi ---
fig2 = px.histogram(df, x="Glucose_Risk", color="Outcome", barmode="group",
                    title="Glikoz Risk Kategorisine Göre Diyabet Dağılımı",
                    color_discrete_sequence=['#3B82F6', '#EF4444'])
fig2.update_layout(paper_bgcolor='#0F172A', plot_bgcolor='#0F172A', font_color='#CBD5E1')
fig2.show()
fig2.write_html("figures/glucose_risk_distribution.html")

# 4. Kategorik Değişkenleri Encode Etme (One-Hot Encoding)
# Makine öğrenmesi algoritmaları metin verisi (Normal, Obese vb.) anlayamaz.
df_processed = pd.get_dummies(df, columns=['Glucose_Risk', 'BMI_Category', 'Age_Group'], drop_first=True)

# Sütun isimlerindeki potansiyel boşlukları veya sorunlu karakterleri temizleyelim (XGBoost gibi algoritmalar kızabilir)
df_processed.columns = df_processed.columns.str.replace(' ', '_').str.replace('-', '_')

# 5. İşlenmiş Veriyi data/processed/ klasörüne kaydetme
processed_file_path = 'data/processed/diabetes_processed.csv'
df_processed.to_csv(processed_file_path, index=False)

print(f"\n✅ Veri Ön İşleme Tamamlandı! İşlenmiş ve yeni öznitelikler eklenmiş veri seti şu yola kaydedildi: '{processed_file_path}'")
print(f"📊 Yeni Veri Seti Boyutu: {df_processed.shape[0]} Gözlem, {df_processed.shape[1]} Değişken")
display(df_processed.head())
```

            Makine öğrenmesinde aykırı değerler genellikle model performansını düşürdüğü için veri setinden silinir (Drop). Ancak <b>tıbbi veri setlerinde çok yüksek İnsülin veya çok yüksek Vücut Kitle İndeksi (BMI) genellikle bir veri giriş hatası değil, hastanın gerçek ve kritik bir patolojik durumudur.</b> Bu projede aykırı değerleri agresif bir biçimde silmek (trimming) yerine, bu değerlere karşı dirençli olan (Robust) Ağaç tabanlı algoritmaları (Random Forest, XGBoost) tercih etme stratejisi benimsenecektir.

```python
import plotly.figure_factory as ff
import plotly.express as px
import plotly.graph_objects as go

print("📊 İleri Düzey Görselleştirmeler Oluşturuluyor...\n")

# Yorumlanabilirliği artırmak için Outcome değişkenini string yapalım (Sadece grafikler için)
df_plot = df.copy() # df, bir önceki adımdan gelen (encoded OLMAYAN) verimiz.
df_plot['Outcome_Label'] = df_plot['Outcome'].map({0: 'Sağlıklı (0)', 1: 'Diyabetik (1)'})

# --- GÖRSELLEŞTİRME 3: Korelasyon Isı Haritası (Correlation Heatmap) ---
# Sadece sayısal kolonları seçip korelasyonu hesaplayalım
numeric_cols = ['Pregnancies', 'Glucose', 'BloodPressure', 'SkinThickness', 'Insulin', 'BMI', 'DiabetesPedigreeFunction', 'Age', 'Outcome']
corr_matrix = df_plot[numeric_cols].corr().round(2)

fig3 = px.imshow(corr_matrix, text_auto=True, aspect="auto",
                 color_continuous_scale='RdBu_r', zmin=-1, zmax=1,
                 title="Öznitelikler Arası Korelasyon Isı Haritası (Pearson)")
fig3.update_layout(paper_bgcolor='#0F172A', plot_bgcolor='#0F172A', font_color='#CBD5E1')
fig3.show()
fig3.write_html("figures/correlation_heatmap.html")

# --- GÖRSELLEŞTİRME 4: Keman Grafiği (Violin Plot) - Glikoz Yoğunluğu ---
# Hangi glikoz seviyelerinde sağlıklı ve hastalar ayrışıyor?
fig4 = px.violin(df_plot, y="Glucose", x="Outcome_Label", color="Outcome_Label", 
                 box=True, points="all", hover_data=df_plot.columns,
                 color_discrete_sequence=['#3B82F6', '#EF4444'],
                 title="Diyabet Durumuna Göre Glikoz Dağılımı ve Yoğunluğu")
fig4.update_layout(paper_bgcolor='#0F172A', plot_bgcolor='#0F172A', font_color='#CBD5E1', showlegend=False)
fig4.show()
fig4.write_html("figures/glucose_violin_plot.html")

# --- GÖRSELLEŞTİRME 5: Kutu Grafiği (Box Plot) - Yaş ve BMI Kesişimi ---
# Aykırı değerleri (outliers) ve medyanları net görmek için
fig5 = px.box(df_plot, x="Outcome_Label", y="BMI", color="Outcome_Label",
              color_discrete_sequence=['#3B82F6', '#EF4444'],
              title="Vücut Kitle İndeksi (BMI) Aykırı Değer (Outlier) Analizi")
fig5.update_layout(paper_bgcolor='#0F172A', plot_bgcolor='#0F172A', font_color='#CBD5E1', showlegend=False)
fig5.show()
fig5.write_html("figures/bmi_boxplot.html")

# --- GÖRSELLEŞTİRME 6: 3 Boyutlu Dağılım (3D Scatter Plot) ---
# Hastalığı en çok etkileyen 3 değişkeni 3 boyutlu uzayda görelim
fig6 = px.scatter_3d(df_plot, x='Glucose', y='BMI', z='Age', color='Outcome_Label',
                     color_discrete_sequence=['#3B82F6', '#EF4444'], opacity=0.7,
                     title="Glucose, BMI ve Yaşın 3 Boyutlu Uzayda Dağılımı")
fig6.update_layout(paper_bgcolor='#0F172A', plot_bgcolor='#0F172A', font_color='#CBD5E1',
                   scene=dict(xaxis=dict(backgroundcolor='#1E293B'),
                              yaxis=dict(backgroundcolor='#1E293B'),
                              zaxis=dict(backgroundcolor='#1E293B')))
fig6.update_traces(marker=dict(size=4))
fig6.show()
fig6.write_html("figures/3d_scatter_glucose_bmi_age.html")

print("✅ Grafikler başarıyla oluşturuldu ve 'figures/' klasörüne HTML olarak kaydedildi.")
print("💡 DİKKAT: Isı haritasında Glikoz ile Outcome arasında (0.50) seviyesinde güçlü bir pozitif ilişki görülmektedir.")
```

Bu projede, verinin altındaki hem doğrusal (linear) hem de karmaşık/doğrusal olmayan (non-linear) örüntüleri yakalayabilmek için 4 farklı öğrenme paradigmasından toplam <b>10 farklı sınıflandırma algoritması</b> seçilmiştir:

        <ul style="color: #CBD5E1; font-size: 14px; line-height: 1.6; font-family: 'Arial', sans-serif;">
            <li><b>İstatistiksel / Olasılıksal Modeller:</b> Logistic Regression, Gaussian Naive Bayes</li>
            <li><b>Uzaklık / Marj Tabanlı Modeller:</b> K-Nearest Neighbors (KNN), Support Vector Machines (SVM)</li>
            <li><b>Ağaç Tabanlı (Tree-Based) Modeller:</b> Decision Tree</li>
            <li><b>Topluluk (Ensemble) - Bagging & Boosting:</b> Random Forest, Extra Trees, AdaBoost, Gradient Boosting, XGBoost</li>
        </ul>
        <h4 style="color: #C4B5FD; margin-bottom: 5px; margin-top: 20px;">⚙️ Model Eğitim Mimarisi (Validation Strategy)</h4>
        <p style="color: #CBD5E1; font-size: 15px; line-height: 1.6; font-family: 'Arial', sans-serif;">
            <b>1. Train-Test Split (Veri Bölme):</b> Veri seti %80 Eğitim ve %20 Test olacak şekilde ayrılmıştır. Sınıf dengesizliğini korumak için <code>stratify=y</code> parametresi kullanılmış, böylece hem eğitim hem de test setinde Diyabet/Sağlıklı oranının aynı kalması sağlanmıştır.<br>
            <b>2. Data Scaling (Standardizasyon):</b> Uzaklık tabanlı (KNN, SVM) ve Gradyan İnişi kullanan (Logistic Reg.) algoritmaların yüksek varyanslı özelliklerden (örn: İnsülin=500, Pedigree=0.2) etkilenmemesi için bağımsız değişkenler <i>StandardScaler</i> ile Z-Skoruna dönüştürülmüştür. Ağaç tabanlı algoritmalar ölçeklemeden etkilenmese de, tüm modellere standart bir veri beslemek metodolojik bir standarttır.<br>
            <b>3. Cross-Validation:</b> Modellerin sadece şans eseri iyi bir eğitim setine denk gelip ezber yapmasını (Overfitting) önlemek için <b>5-Fold Stratified Cross Validation</b> uygulanmış ve modelin stabilitesi test edilmiştir.
        </p>
        <h4 style="color: #F87171; margin-bottom: 5px; margin-top: 20px;">⚖️ Değerlendirme Metrikleri: Objektif Kıyaslama</h4>
        <p style="color: #CBD5E1; font-size: 15px; line-height: 1.6; font-family: 'Arial', sans-serif;">
            Bu aşamada ön yargılı bir model seçimi yapılmayacaktır. Algoritmalar; <b>Accuracy (Doğruluk)</b>, <b>Precision (Kesinlik)</b>, <b>Recall (Duyarlılık - Hastayı Kaçırmama)</b>, <b>F1-Score</b> ve <b>ROC-AUC (Sınıf Ayırma Gücü)</b> metrikleri üzerinden tam bağımsız olarak test edilecek ve sonuçlar analitik grafiklerle incelenecektir.
        </p>
    </div>
</div>

    <p style="color: #CBD5E1; font-size: 14px; line-height: 1.6; font-family: 'Arial', sans-serif;">
        Medikal veriler, doğrusal ilişkiler (örn: kilo artarsa tansiyon artar) ile doğrusal olmayan karmaşık ilişkilerin (örn: belli bir yaş ve insülin seviyesi kombinasyonunun diyabeti tetiklemesi) bir karışımıdır. Bu karmaşayı çözebilmek için 4 farklı matematiksel paradigmaya ait algoritmalar bilinçli olarak seçilmiştir:
    </p>
    <div style="display: flex; flex-wrap: wrap; gap: 15px; margin-top: 15px;">        
        <!-- Kutu 1: İstatistiksel -->
        <div style="flex: 1 1 45%; background-color: #0F172A; padding: 15px; border-radius: 8px;">
            <h4 style="color: #60A5FA; margin-top: 0; margin-bottom: 5px;">1. İstatistiksel ve Olasılıksal</h4>
            <p style="color: #94A3B8; font-size: 13px; margin: 0;">
                <b>Logistic Regression:</b> Tıp literatüründe altın standarttır. Hangi özniteliğin hastalığı ne yönde (katsayılarla) etkilediğini en şeffaf anlatan temel "Baseline" modeldir.<br><br>
                <b>Gaussian Naive Bayes:</b> Bayes teoremine dayanır. Semptomlara bakarak hastalığın "olma olasılığını" hesaplar. Sağlık alanında düşük False Negative (atlanan hasta) oranıyla meşhurdur.
            </p>
        </div>
        <!-- Kutu 2: Uzay / Uzaklık -->
        <div style="flex: 1 1 45%; background-color: #0F172A; padding: 15px; border-radius: 8px;">
            <h4 style="color: #34D399; margin-top: 0; margin-bottom: 5px;">2. Uzay ve Marj Tabanlı</h4>
            <p style="color: #94A3B8; font-size: 13px; margin: 0;">
                <b>KNN (K-Nearest Neighbors):</b> Hastaları biyolojik benzerliklerine (Öklid uzaklığı) göre gruplar. "Senin kan değerlerin diyabetik hastalara daha çok benziyor" mantığıyla çalışır.<br><br>
                <b>SVM (Support Vector Machine):</b> Sınıflar (Sağlıklı vs Hasta) iç içe geçmişse, kernel trick (çekirdek hilesi) kullanarak veriyi çok boyutlu uzaya çeker ve aralarından maksimum genişlikte bir sınır çizer.
            </p>
        </div>
        <!-- Kutu 3: Kural Tabanlı -->
        <div style="flex: 1 1 45%; background-color: #0F172A; padding: 15px; border-radius: 8px;">
            <h4 style="color: #FBBF24; margin-top: 0; margin-bottom: 5px;">3. Ağaç / Kural Tabanlı</h4>
            <p style="color: #94A3B8; font-size: 13px; margin: 0;">
                <b>Decision Tree:</b> İnsan zihnine (ve bir doktorun triyaj sürecine) en yakın algoritmadır. "Glikoz 140'tan büyük mü? Evet ise Yaş 40'tan büyük mü?" şeklinde kural dizileri çıkarır.
            </p>
        </div>
        <!-- Kutu 4: Topluluk (Ensemble) -->
        <div style="flex: 1 1 45%; background-color: #0F172A; padding: 15px; border-radius: 8px;">
            <h4 style="color: #F87171; margin-top: 0; margin-bottom: 5px;">4. Ensemble (Topluluk) Modelleri</h4>
            <p style="color: #94A3B8; font-size: 13px; margin: 0;">
                <i>Burada amaç zayıf modellerin hatalarını birleştirerek güçlü bir "Kurul (Board of Doctors)" oluşturmaktır.</i><br>
                <b>Random Forest & Extra Trees (Bagging):</b> Yüzlerce karar ağacının oylamasıyla çalışır, tek bir ağacın ezberleme (overfitting) riskini sıfırlar.<br>
                <b>AdaBoost, Gradient Boosting & XGBoost (Boosting):</b> Bir önceki modelin "yanlış teşhis koyduğu" hastalara odaklanarak adım adım kendini iyileştiren, makine öğrenmesinin en agresif ve başarılı algoritmalarıdır.
            </p>
        </div>

```python
import pandas as pd
import numpy as np
import plotly.express as px
import plotly.graph_objects as go
from sklearn.model_selection import train_test_split, cross_val_score, StratifiedKFold
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, roc_auc_score
import warnings
warnings.filterwarnings('ignore')

# 1. İşlenmiş Veriyi Yükleme
df_processed = pd.read_csv('data/processed/diabetes_processed.csv')

# 2. Bağımsız(X) ve Bağımlı(y) Değişkenleri Ayırma
X = df_processed.drop('Outcome', axis=1)
y = df_processed['Outcome']

# 3. %80 Eğitim, %20 Test Ayrımı (Stratified - Sınıf oranlarını koruyarak)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.20, random_state=42, stratify=y)

# 4. Standardizasyon (Z-Score Normalization)
# Fit işlemini SADECE Eğitim setine yapıyoruz ki, test setinden veri sızması (Data Leakage) olmasın!
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

# 5. Modellerin Kütüphanelerini Yükleme ve Tanımlama
from sklearn.linear_model import LogisticRegression
from sklearn.naive_bayes import GaussianNB
from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import SVC
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier, ExtraTreesClassifier, AdaBoostClassifier, GradientBoostingClassifier
from xgboost import XGBClassifier

models = {
    "Logistic Regression": LogisticRegression(max_iter=1000, random_state=42),
    "Naive Bayes": GaussianNB(),
    "KNN": KNeighborsClassifier(n_neighbors=5),
    "SVM": SVC(probability=True, kernel='rbf', random_state=42),
    "Decision Tree": DecisionTreeClassifier(max_depth=5, random_state=42),
    "Random Forest": RandomForestClassifier(n_estimators=100, random_state=42),
    "Extra Trees": ExtraTreesClassifier(n_estimators=100, random_state=42),
    "AdaBoost": AdaBoostClassifier(random_state=42),
    "Gradient Boosting": GradientBoostingClassifier(random_state=42),
    "XGBoost": XGBClassifier(use_label_encoder=False, eval_metric='logloss', random_state=42)
}

print("⏳ 10 Farklı Model Eğitiliyor ve Çapraz Doğrulamaları Yapılıyor...\n")

# Sonuçları tutacağımız liste
results = []

# Stratified K-Fold ayarı
cv = StratifiedKFold(n_splits=5, shuffle=True, random_state=42)

for name, model in models.items():
    # A. Cross Validation (Eğitim seti üzerinde model stabilitesini ölçme)
    cv_acc = cross_val_score(model, X_train_scaled, y_train, cv=cv, scoring='accuracy').mean()
    
    # B. Modeli test için eğitme
    model.fit(X_train_scaled, y_train)
    
    # C. Test seti tahminleri
    y_pred = model.predict(X_test_scaled)
    y_prob = model.predict_proba(X_test_scaled)[:, 1] if hasattr(model, "predict_proba") else [0]*len(y_test)
    
    # D. Metrik Hesaplamaları
    acc = accuracy_score(y_test, y_pred)
    prec = precision_score(y_test, y_pred)
    rec = recall_score(y_test, y_pred)
    f1 = f1_score(y_test, y_pred)
    roc_auc = roc_auc_score(y_test, y_prob)
    
    # Sonuçları kaydet
    results.append({
        "Model": name,
        "CV_Accuracy": cv_acc,
        "Test_Accuracy": acc,
        "Precision": prec,
        "Recall": rec,
        "F1_Score": f1,
        "ROC_AUC": roc_auc
    })

# Pandas DataFrame'e Çevirme
df_results = pd.DataFrame(results).round(3)

print("✅ Modellerin Eğitimi Tamamlandı! İşte Objektif Performans Tablosu:")
display(df_results.sort_values(by="Recall", ascending=False).reset_index(drop=True))

# --- GÖRSELLEŞTİRME 7: Recall Performansları Bar Grafiği ---
# Sağlık verisinde "Hastayı Atlamama (Recall)" çok kritik olduğu için önce buna bakıyoruz.
fig7 = px.bar(df_results.sort_values('Recall', ascending=False), 
              x='Model', y='Recall', color='Recall',
              title='10 Algoritmanın Recall (Gerçek Hastaları Yakalama) Performansları',
              text='Recall', color_continuous_scale='Turbo')
fig7.update_layout(paper_bgcolor='#0F172A', plot_bgcolor='#0F172A', font_color='#CBD5E1')
fig7.update_traces(texttemplate='%{text:.3f}', textposition='outside')
fig7.show()
fig7.write_html("figures/models_recall_comparison.html")

# --- GÖRSELLEŞTİRME 8: Çok Eksenli Radar (Spider) Grafiği ---
# Modellerin trade-off'larını (Örn: Recall yüksek ama Precision düşük mü?) görmek için şahane bir grafik.
fig8 = go.Figure()

for i, row in df_results.iterrows():
    fig8.add_trace(go.Scatterpolar(
        r=[row['Test_Accuracy'], row['Precision'], row['Recall'], row['F1_Score'], row['ROC_AUC']],
        theta=['Accuracy', 'Precision', 'Recall', 'F1_Score', 'ROC-AUC'],
        fill='toself',
        name=row['Model'],
        visible='legendonly' # Sadece legend'a tıklanarak açılsın ki karmaşa olmasın (Interaktif özellik)
    ))

# Varsayılan olarak XGBoost, Naive Bayes ve Random Forest açık gelsin
fig8.data[1].visible = True # Naive Bayes
fig8.data[5].visible = True # Random Forest
fig8.data[9].visible = True # XGBoost

fig8.update_layout(
    polar=dict(
        radialaxis=dict(visible=True, range=[0.5, 1.0], gridcolor='#334155'),
        bgcolor='#1E293B'
    ),
    paper_bgcolor='#0F172A',
    font_color='#CBD5E1',
    title="Çok Eksenli (Radar) Algoritma Performans Karşılaştırması<br><span style='font-size:12px;'><i>İncelemek istediğiniz modeli yandaki efsaneden (legend) açıp kapatabilirsiniz.</i></span>"
)
fig8.show()
fig8.write_html("figures/models_radar_chart.html")
```

            <span style="font-size: 16px; color: #FCA5A5;">(CRISP-DM Faz 4 & 5: Model Tuning & Evaluation)</span>
        </h2>        
        <h4 style="color: #F87171; margin-bottom: 5px;">🏆 Neden Bu 3 Model Seçildi?</h4>
        <p style="color: #CBD5E1; font-size: 15px; line-height: 1.6; font-family: 'Arial', sans-serif;">
            Baseline test sonuçlarına dayanarak, sağlık Karar Destek Sistemleri (CDSS) mimarisine en uygun olan <b>Random Forest, Gradient Boosting ve XGBoost</b> modelleri seçilmiştir. Bu modellerin ortak özelliği <b>"Ağaç Tabanlı Topluluk (Tree-Based Ensemble)"</b> algoritmaları olmalarıdır. Biyolojik verilerdeki aykırı gözlemlere (Örn: Aşırı obezite veya çok yüksek insülin seviyeleri) karşı dirençlidirler (Robust) ve verinin karmaşık yapısındaki doğrusal olmayan ilişkileri çözmede istatistiksel modellere göre çok daha üstündürler.

        <h4 style="color: #F87171; margin-bottom: 5px; margin-top: 20px;">🛠️ GridSearchCV: "İyi" Modeli "Mükemmel" Yapmak</h4>
        <p style="color: #CBD5E1; font-size: 15px; line-height: 1.6; font-family: 'Arial', sans-serif;">
            Modellerin varsayılan (default) parametreleri veri setimize tam oturmayabilir (Underfitting veya Overfitting). Bu nedenle hiperparametre uzayında sistematik bir arama yapılacaktır:
            <ul style="color: #CBD5E1; font-size: 15px; line-height: 1.6;">
                <li><b>Ağaç Sayısı ve Derinliği:</b> Ağaçların (Estimator) sayısı artırılırken, derinlikleri (Max Depth) sınırlandırılarak modelin ezber yapması engellenecektir.</li>
                <li><b>Öğrenme Oranı (Learning Rate):</b> Boosting modelleri için hata düzeltme adımlarının büyüklüğü optimize edilecektir.</li>
                <li><b>Optimizasyon Hedefi (Scoring):</b> Tıbbi veri setlerindeki hayati önemi sebebiyle GridSearchCV, doğrudan 'Accuracy' yerine <b>'Recall'</b> (Hassasiyet) metriğini maksimize edecek şekilde yönlendirilecektir.</li>
            </ul>
        </p>
    </div>
</div>

```python
from sklearn.model_selection import GridSearchCV
import time

print("⚙️ Modeller için Hiperparametre Optimizasyonu (GridSearchCV) Başlatıldı...\n")
print("Hedef Metrik: RECALL (Hastaları Kaçırmamak)")
print("Bu işlem biraz zaman alabilir, lütfen bekleyiniz...\n")

# 1. Optimize Edilecek Modeller ve Parametre Uzayları
tuning_models = {
    
    "Random Forest": {
        "model": RandomForestClassifier(random_state=42),
        "params": {
            'n_estimators': [50, 100, 200],
            'max_depth': [None, 5, 10, 15],
            'min_samples_split': [2, 5, 10],
            'class_weight': ['balanced', None] # Azınlık sınıfı (Diyabet) için ağırlıklandırma
        }
    },
    
    "Gradient Boosting": {
        "model": GradientBoostingClassifier(random_state=42),
        "params": {
            'n_estimators': [50, 100, 200],
            'learning_rate': [0.01, 0.1, 0.2],
            'max_depth': [3, 5, 7]
        }
    },
    
    "XGBoost": {
        "model": XGBClassifier(use_label_encoder=False, eval_metric='logloss', random_state=42),
        "params": {
            'n_estimators': [50, 100, 200],
            'learning_rate': [0.01, 0.1, 0.2],
            'max_depth': [3, 5, 7],
            'scale_pos_weight': [1, 1.5, 2] # Sınıf dengesizliğini çözmek için (Hasta/Sağlıklı oranı)
        }
    }
}

# 2. Optimizasyon Süreci
tuned_results = []
best_estimators = {}

for name, mp in tuning_models.items():
    start_time = time.time()
    
    # GridSearch Tanımlaması (Recall'a göre optimize ediliyor)
    clf = GridSearchCV(mp["model"], mp["params"], cv=5, scoring='recall', n_jobs=-1)
    clf.fit(X_train_scaled, y_train)
    
    # En iyi modeli test setinde değerlendirme
    best_model = clf.best_estimator_
    best_estimators[name] = best_model
    
    y_pred = best_model.predict(X_test_scaled)
    y_prob = best_model.predict_proba(X_test_scaled)[:, 1]
    
    acc = accuracy_score(y_test, y_pred)
    prec = precision_score(y_test, y_pred)
    rec = recall_score(y_test, y_pred)
    f1 = f1_score(y_test, y_pred)
    roc_auc = roc_auc_score(y_test, y_prob)
    
    end_time = time.time()
    
    tuned_results.append({
        "Model": name,
        "Best_Params": str(clf.best_params_),
        "Tuned_Accuracy": acc,
        "Tuned_Precision": prec,
        "Tuned_Recall": rec,
        "Tuned_F1": f1,
        "Tuned_ROC_AUC": roc_auc,
        "Eğitim Süresi (Sn)": round(end_time - start_time, 2)
    })
    print(f"✅ {name} Optimizasyonu Tamamlandı. (Süre: {round(end_time - start_time, 2)} sn)")

# 3. Optimize Edilmiş Sonuçları Tablolama
df_tuned_results = pd.DataFrame(tuned_results)
df_display = df_tuned_results.drop(columns=['Best_Params']) # Tablo karmaşık olmasın diye çıkardık
print("\n🏆 GridSearch Sonrası Optimize Edilmiş Model Performansları (Recall Odaklı):")
display(df_display.sort_values(by="Tuned_Recall", ascending=False).reset_index(drop=True))

# 4. GÖRSELLEŞTİRME 9: Baseline (Optimizasyon Öncesi) ve Tuned (Optimizasyon Sonrası) Recall Karşılaştırması
# Eski sonuçlardan bu 3 modelin değerlerini çekelim
old_recalls = df_results[df_results['Model'].isin(["Random Forest", "Gradient Boosting", "XGBoost"])][['Model', 'Recall']]
new_recalls = df_tuned_results[['Model', 'Tuned_Recall']]
comparison_df = pd.merge(old_recalls, new_recalls, on='Model')

import plotly.graph_objects as go
fig9 = go.Figure(data=[
    go.Bar(name='Optimizasyon Öncesi (Default)', x=comparison_df['Model'], y=comparison_df['Recall'], marker_color='#3B82F6'),
    go.Bar(name='Optimizasyon Sonrası (Tuned)', x=comparison_df['Model'], y=comparison_df['Tuned_Recall'], marker_color='#10B981')
])

fig9.update_layout(barmode='group', title='GridSearchCV Optimizasyonu: Recall Performansındaki Gelişim',
                   paper_bgcolor='#0F172A', plot_bgcolor='#0F172A', font_color='#CBD5E1',
                   yaxis_title='Recall Skoru (Hastayı Yakalama Oranı)')
fig9.show()
fig9.write_html("figures/tuning_recall_comparison.html")
```

Eger bir model eğitim setinde %95 üstü performans gösterip test setinde %75'lerde kalıyorsa, bu model <b>Overfitting (Aşırı Öğrenme)</b> yaşıyor demektir ve klinik kullanıma uygun değildir.

                <li><b>XGBoost:</b> L2 Regularization (reg_lambda) ve Gamma (ağaç dallanma cezası) parametreleri yükseltilerek modelin agresif öğrenme yapısı yavaşlatılacaktır.</li>

    </div>

```python
print("🔍 Modellerin Eğitim (Train) ve Test Skorları Karşılaştırılıyor...\n")

overfit_check = []

# best_estimators bir önceki hücreden (Hücre 10) geliyor
for name, model in best_estimators.items():
    # Eğitim Seti Tahminleri
    y_train_pred = model.predict(X_train_scaled)
    train_rec = recall_score(y_train, y_train_pred)
    train_acc = accuracy_score(y_train, y_train_pred)
    
    # Test Seti Tahminleri
    y_test_pred = model.predict(X_test_scaled)
    test_rec = recall_score(y_test, y_test_pred)
    test_acc = accuracy_score(y_test, y_test_pred)
    
    overfit_check.append({
        "Model": name,
        "Train_Recall": train_rec,
        "Test_Recall": test_rec,
        "Recall_Farkı (Gap)": train_rec - test_rec,
        "Train_Accuracy": train_acc,
        "Test_Accuracy": test_acc
    })

df_overfit = pd.DataFrame(overfit_check).round(3)
display(df_overfit)

# --- GÖRSELLEŞTİRME 10: Overfitting Analizi Bar Grafiği ---
fig10 = go.Figure(data=[
    go.Bar(name='Eğitim (Train) Recall', x=df_overfit['Model'], y=df_overfit['Train_Recall'], marker_color='#8B5CF6'),
    go.Bar(name='Test Recall', x=df_overfit['Model'], y=df_overfit['Test_Recall'], marker_color='#EC4899')
])
fig10.update_layout(barmode='group', title='Overfitting Tespiti: Eğitim vs Test Recall Skorları',
                   paper_bgcolor='#0F172A', plot_bgcolor='#0F172A', font_color='#CBD5E1')
fig10.show()
fig10.write_html("figures/overfitting_check.html")

print("\n⚠️ YORUM: Eğer yukarıdaki tabloda Train_Recall 0.95-1.00 civarında iken, Test_Recall 0.70'lerde ise ciddi bir OVERFITTING vardır.")
print("🛠️ Modelleri 'Düzenlileştirme (Regularization)' ve 'Budama (Pruning)' ile tekrar eğitiyoruz...\n")

# --- EZBERİ BOZMAK İÇİN SIKI KISITLAMALI (ROBUST) YENİ PARAMETRELER ---
robust_tuning_models = {
    "Random Forest (Robust)": {
        "model": RandomForestClassifier(random_state=42),
        "params": {
            'n_estimators': [100],
            'max_depth': [3, 5], # Derinliği agresif şekilde kestik
            'min_samples_leaf': [5, 10], # Yapraklarda en az 5-10 hasta kalmalı ki ezber yapamasın
            'class_weight': ['balanced']
        }
    },
    "Gradient Boosting (Robust)": {
        "model": GradientBoostingClassifier(random_state=42),
        "params": {
            'n_estimators': [100],
            'learning_rate': [0.05], # Yavaş öğrenme
            'max_depth': [3], # Sığ ağaçlar
            'subsample': [0.8] # Verinin sadece %80'ini kullanarak rassallık katma
        }
    },
    "XGBoost (Robust)": {
        "model": XGBClassifier(use_label_encoder=False, eval_metric='logloss', random_state=42),
        "params": {
            'n_estimators': [100],
            'learning_rate': [0.05],
            'max_depth': [3],
            'gamma': [1, 5], # Yüksek gamma = Daha az dallanma (Cezalandırma)
            'reg_lambda': [10, 50], # L2 Cezası çok artırıldı
            'scale_pos_weight': [1.5]
        }
    }
}

final_robust_estimators = {}
robust_results = []

for name, mp in robust_tuning_models.items():
    clf = GridSearchCV(mp["model"], mp["params"], cv=5, scoring='recall', n_jobs=-1)
    clf.fit(X_train_scaled, y_train)
    
    best_model = clf.best_estimator_
    final_robust_estimators[name] = best_model
    
    # Yeni Train / Test Skorları
    train_rec = recall_score(y_train, best_model.predict(X_train_scaled))
    test_rec = recall_score(y_test, best_model.predict(X_test_scaled))
    test_acc = accuracy_score(y_test, best_model.predict(X_test_scaled))
    test_auc = roc_auc_score(y_test, best_model.predict_proba(X_test_scaled)[:, 1])
    
    robust_results.append({
        "Model": name,
        "Train_Recall": train_rec,
        "Test_Recall": test_rec,
        "Fark (Gap)": train_rec - test_rec,
        "Final_Test_Accuracy": test_acc,
        "Final_ROC_AUC": test_auc
    })

df_robust = pd.DataFrame(robust_results).round(3)
print("✅ Düzenlileştirme (Regularization) Sonrası Final Performanslar:")
display(df_robust)  
```

```python
from plotly.subplots import make_subplots
import plotly.graph_objects as go
from sklearn.metrics import confusion_matrix, roc_curve, auc

print("📊 Top 3 Modelin Karmaşıklık Matrisleri ve ROC Eğrileri Hazırlanıyor...\n")

# Modellerin isimlerindeki '(Robust)' takısını atarak başlıkları sadeleştirelim
clean_model_names = [name.replace(" (Robust)", "") for name in final_robust_estimators.keys()]

# --- GÖRSELLEŞTİRME 11: 3 Modelin Yan Yana Confusion Matrix Karşılaştırması ---
# horizontal_spacing artırıldı, grafik ferahlatıldı.
fig_cm_comp = make_subplots(rows=1, cols=3, 
                            subplot_titles=clean_model_names, 
                            horizontal_spacing=0.15)

# Kısa ve öz etiketler (Kesişmeleri engellemek için)
x_labels = ['Sağlıklı', 'Diyabetik']
y_labels = ['Diyabetik', 'Sağlıklı'] # Plotly aşağıdan yukarı çizdiği için y ekseni ters yazılır

for i, (name, model) in enumerate(final_robust_estimators.items(), start=1):
    y_pred = model.predict(X_test_scaled)
    cm = confusion_matrix(y_test, y_pred)
    
    # Isı haritası oluşturma
    heatmap = go.Heatmap(
        z=cm[::-1], # Ekseni düzeltmek için
        text=cm[::-1], 
        texttemplate="<b>%{text}</b>", # Kalın ve belirgin font
        textfont={"size": 22, "color": "white", "family": "Arial"},
        colorscale='Teal', 
        showscale=False,
        x=x_labels,
        y=y_labels,
        hoverinfo="skip" # Üzerine gelince çıkan gereksiz kutucukları kapattık
    )
    fig_cm_comp.add_trace(heatmap, row=1, col=i)
    
    # Sadece en soldaki grafiğe Y ekseni (Gerçek Değer) başlığı ekleyelim ki kalabalık olmasın
    if i == 1:
        fig_cm_comp.update_yaxes(title_text="<b>Gerçek Durum</b>", row=1, col=i, title_font=dict(color="#CBD5E1"))
    
    # Tüm alt grafiklere X ekseni (Tahmin) başlığı ekleyelim
    fig_cm_comp.update_xaxes(title_text="<b>Tahmin</b>", row=1, col=i, title_font=dict(color="#CBD5E1"))

# Genel düzen (Layout) ayarları
fig_cm_comp.update_layout(
    title_text="<b>Klinik Kıyaslama: Karmaşıklık Matrisleri (False Negative Sayılarına Odaklanın)</b>",
    title_font=dict(size=20, color="#CFFAFE"),
    paper_bgcolor='#0F172A', 
    plot_bgcolor='#0F172A', 
    font_color='#94A3B8',
    width=1100,  # Grafiği genişleterek iç içe girmeyi önledik
    height=450,  # Orantılı yükseklik
    margin=dict(t=100, b=50, l=50, r=50) # Kenar boşlukları
)

# Alt başlıkların renk ve boyut ayarı
for annotation in fig_cm_comp['layout']['annotations']:
    annotation['font'] = dict(size=16, color='#38BDF8')

fig_cm_comp.show()
fig_cm_comp.write_html("figures/top3_confusion_matrices.html")


# --- GÖRSELLEŞTİRME 12: 3 Modelin Kombine ROC Eğrisi ---
fig_roc_comp = go.Figure()
colors = ['#38BDF8', '#10B981', '#F43F5E'] # Mavi, Yeşil, Kırmızı

for (name, model), color in zip(final_robust_estimators.items(), colors):
    y_prob = model.predict_proba(X_test_scaled)[:, 1]
    fpr, tpr, _ = roc_curve(y_test, y_prob)
    auc_val = auc(fpr, tpr)
    
    clean_name = name.replace(" (Robust)", "")
    
    fig_roc_comp.add_trace(go.Scatter(
        x=fpr, y=tpr, mode='lines', 
        name=f"<b>{clean_name}</b> (AUC={auc_val:.3f})", 
        line=dict(color=color, width=3.5)
    ))

# Rassal tahmin çizgisi
fig_roc_comp.add_trace(go.Scatter(x=[0, 1], y=[0, 1], mode='lines', name='Rassal Tahmin',
                             line=dict(color='#64748B', width=2, dash='dash')))

fig_roc_comp.update_layout(
    title="<b>Kombine ROC Eğrisi: Sınıfları Ayırt Etme Gücü (AUC Kıyaslaması)</b>",
    title_font=dict(size=20, color="#CFFAFE"),
    xaxis_title="False Positive Rate (Yanlış Alarm Oranı)",
    yaxis_title="True Positive Rate (Doğru Teşhis / Recall)",
    paper_bgcolor='#0F172A', 
    plot_bgcolor='#0F172A', 
    font_color='#CBD5E1',
    width=900,
    height=600,
    legend=dict(x=0.65, y=0.1, bgcolor='rgba(15, 23, 42, 0.8)', bordercolor='#334155', borderwidth=1)
)
fig_roc_comp.show()
fig_roc_comp.write_html("figures/top3_roc_curves.html")
```
